```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

## 折线图

```python
x = np.linspace(0, 10, 100)
x
```

```python
siny = np.sin(x)
siny
```

```python
plt.plot(x, siny)
```

```python
cosy = np.cos(x)
```

```python
plt.plot(x, cosy)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='--')
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='--')
# x轴显示范围
plt.xlim(-5, 15)
# y轴显示范围
plt.ylim(-5, 5)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='--')
# 也可以通过axis来设置轴的取值范围
plt.axis([-5, 15, -5, 5])
```

```python
plt.plot(x, siny, label='sin(x)')
plt.plot(x, cosy, color='red', linestyle='--', label='sin(y)')
# 也可以通过axis来设置轴的取值范围
plt.axis([-5, 15, -5, 5])
plt.xlabel('x axis')
plt.ylabel('y axis')
# 开启视图
plt.legend()
plt.title('hello world')
```

## 散点图

```python
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
```

```python
def show(x, y, alpha=0.5):
    plt.scatter(x, y, alpha=alpha) #点的透明度为0.5
    # 也可以通过axis来设置轴的取值范围
    plt.xlabel('x axis')
    plt.ylabel('y axis')
    plt.title('hello world')
```

```python
show(x, y)
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
```

```python
show(x, y, 0.1)
```
